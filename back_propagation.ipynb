```python
#BackPropagation Algorithm 
####### Download iris.csv from repo to run the algorithm #########
# In this question also I will use iris dataset , first we will make multilayer network and activation functio
# will be a sigmoid function
#Backpropagation Algorithm is the Algorithm that train given forward neuralnet for the input patterns where the classifications is known to us.The error value is recorded after output response and weights of cells are changed . It consists of mainly three layers : 1)input , 2) hidden ,3)output
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split



# read data set 

dataset = pd.read_csv("iris.csv")
dataset.head()
```

```python
# Lets put labels for these classes as 0,1,2
dataset['Species'].replace(['Iris-setosa','Iris-virginica','Iris-versicolor'],[0,1,2],inplace=True)
```

```python
dataset.head()
```

```python
#Perform one hot encoding to the output 
def one_hot(Y):
  n_col=np.amax(Y)+1
  binarize = np.zeros((len(Y),n_col))
  for i in range(len(Y)):
    binarize[i,Y[i]]=1
  return binarize

# Add sigmoid function 

def sigmoid(x):
  return (1/(np.exp(-x)+1))
def sigmoid_derivative(x):
  return sigmoid(x)*(1-sigmoid(x))

#Adding a normalize func

def normalise(X,axis=-1,order=2):
  A2=np.atleast_1d(np.linalg.norm(X,order,axis))
  A2[A2==0]=1
  return X/np.expand_dims(A2,axis)
```

```python
#Applying normalisation function and encoding to output

columns=['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']
x=dataset
x=normalise(x.values)
print(x)
columns=['Species']
y=pd.DataFrame(dataset,columns=columns)
y=y.values
y=y.flatten()
y=one_hot(y)
print(y)
```

```python
# Split data in test and train set
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.33)


#adding weights
w1 = 2*np.random.random((5,5))-1
w2 = 2*np.random.random((5,3))-1

n=0.1

#making a set for errors to see how it is decreased when network is trained

errors=[]

#Making a feed forward and a back propagation network 

for i in range(100000):
  layer_0=X_train
  layer_1=sigmoid(np.dot(layer_0,w1))
  layer_2=sigmoid(np.dot(layer_1,w2))
  #BackPropagation Network
  layer_2_error=y_train-layer_2
  layer_2_delta=layer_2_error*sigmoid_derivative(layer_2)
  layer_1_error=layer_2_delta.dot(w2.T)
  layer_1_delta=layer_1_error*sigmoid_derivative(layer_1)
  w1+=layer_0.T.dot(layer_1_delta)*n
  w2+=layer_1.T.dot(layer_2_delta)*n
  error=np.mean(np.abs(layer_2_error))
  errors.append(error)
  accuracy_training=(1-error)*100

plt.plot(errors)
plt.xlabel('errors')
plt.ylabel('training data')
plt.show()
```

```python
print('Accuracy'+(str(round(accuracy_training,2))))
```
